현재까지 기온이 가장 높았ㅇ르 때 온도와 날짜는?

```python
import csv
f=open('./seoul.csv')
data=csv.reader(f)
header=next(data)
max_temp=-99999
max_date=''
for row in data:
    try:
        temp=float(row[-1])
        if temp>max_temp:
            max_temp=temp
            max_date=row[0]
    except:
        continue
        
        
print(max_temp, max_date)
f.close()
```
